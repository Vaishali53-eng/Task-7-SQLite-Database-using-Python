```python
import sqlite3
#Step 1: Create and connect to DB
conn = sqlite3.connect('sales_data.db')
cursor = conn.cursor()

#Step 2: Create table
cursor.execute('''
   CREATE TABLE IF NOT EXISTS sales (
    product TEXT,
    quantity INTEGER,
    price INTERGER
    )
''')

#Step 3: Insert fake data
data = [
    ('Shoes', 5, 100),
    ('Shirt', 3, 500),
    ('Shoes', 2, 100),
    ('Jeans', 4, 800),
    ('Shirt', 1, 500)
]
cursor.executemany('INSERT INTO sales VALUES (?, ?, ?)', data)

conn.commit()
conn.close()
```

```python
import sqlite3
import pandas as pd
#Step 4: Connect again
conn = sqlite3.connect('sales_data.db')

# Step 5: SQL query to get summary
query = '''
    SELECT product,
           SUM(quantity) AS total_qty,
           SUM(quantity * price) AS revenue
    FROM sales
    GROUP BY product
'''
#Step 6: Load query result into pandas
df = pd.read_sql_query(query, conn)

#Step 7: Print the result
print(df)

conn.close()
```

```python
import matplotlib.pyplot as plt

#Step 8: plot bar chart
df.plot(kind='bar', x='product', y='revenue')
plt.title('Revenue by Product')
plt.ylabel('Revenue')
plt.xlabel('Product')
plt.tight_layout()

#Optional: Save the chart as image
plt.savefig('sales_chart.png')
plt.show()
```
